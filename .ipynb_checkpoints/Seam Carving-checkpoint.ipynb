## Step 1: Define the energy function

Following the paper (Avidan and Shamir 2007), $e_1$ energy gives good results.

$e_1(I) = |\frac{\partial}{\partial x}I| + |\frac{\partial}{\partial y}I|$ ($I$ represents the image.)

We also noticed that different images can be processed with different energy functions for the optimal results. Here we list another option $e_2$.

$e_2(I) = \sqrt{(\frac{\partial}{\partial x}I)^2 + (\frac{\partial}{\partial y}I)^2}$

```python
def energy(im):
    """
        input: im, H*W*3 (image with 3 color channels)
        output: energy, H*W
        
        Calculates the 2D energy matrix of the image.
    """
    H, W, C = im.shape
    res = np.zeros((H,W)) # stores the results
    
    fil = np.array([-1,0,1]) # filter used to calculate the gradient
    
    for i in range(C):
#         res += (np.abs(cv2.filter2D(im[:,:,i],-1,fil[np.newaxis,:]))+np.abs(cv2.filter2D(im[:,:,i],-1,fil))) # e1
        res += np.sqrt(np.square(cv2.filter2D(im[:,:,i],-1,fil[np.newaxis,:]))+np.square(cv2.filter2D(im[:,:,i],-1,fil))) # e2
    return res
```

## Step 2: Find the minimum seams

First, let's consider the top to bottom case.

We use dynamic programming to solve this problem. We record the total energy cost of the path from the top row to the bottom row. Specifically, we start with the top row where the cost is directly adopted from the energy matrix. The second row records the cumulative sum of the first two rows following the below formula.

$Cost_{i,j} = Energy_{i,j} + min(Cost_{i-1,j-1}, Cost_{i-1,j}, Cost_{i-1,j+1})$

The rows further to the bottom are constructed similarly. In the bottom row, whichever cell has the lowest value represents the last pixel of the optimal vertical seam and the path can be easily traced back.

To avoid the zig-zag effect, for any pixel being removed, we only allow one of the **three nearest top neighbors** to be removed and thus we have j-1, j and j+1 in the above formula.

The left to right case can be constructed very similarly.

```python
# set up dynamic programming matrix

def v_cost(energy):
    """
        input: energy matrix, H*W
        output: 
            cost matrix, H*W. Cumulative energy cost from top to bottom. Any pixel is only allowed to 
                              inherit from one of the three nearest top neighbors.
            path matrix, H*W. Each cell records the RELATIVE column index of the prior pixel in the row above
                              (except for cells in the first row).
        
        Calculates the minimum cumulative energy cost of vertical paths and records the optimal paths.
    """
    H, W = energy.shape
    cost = np.zeros((H,W)) # initialize the cost matrix
    path = np.zeros((H,W),dtype=np.int32) # initialize the path matrix
    
    cost[0,:] = energy[0,:]
    for i in range(1,H):
        for j in range(W):
            if j == 0: # edge case: left-most pixel
                path[i,j] = np.argmin(cost[i-1,j:j+2])
            elif j == W-1: # edge case: right-most pixel
                path[i,j] = np.argmin(cost[i-1,j-1:j+1])-1
            else:
                path[i,j] = np.argmin(cost[i-1,j-1:j+2])-1               
            cost[i,j] = energy[i,j] + cost[i-1,j+path[i,j]]
    return cost, path

def h_cost(energy):
    """
        TODO
    """
    H, W = energy.shape
    cost = np.zeros((H,W)) # initialize the cost matrix
    path = np.zeros((H,W),dtype=np.int32) # initialize the path matrix
    
    cost[:,0] = energy[:,0]
    for j in range(1,W):
        for i in range(H):
            if i == 0: # edge case: top-most pixel
                path[i,j] = np.argmin(cost[i:i+2,j-1])
            elif i == H-1: # edge case: bottom-most pixel
                path[i,j] = np.argmin(cost[i-1:i+1,j-1])-1
            else:
                path[i,j] = np.argmin(cost[i-1:i+2,j-1])-1
            cost[i,j] = energy[i,j] + cost[i+path[i,j],j-1]
    return cost, path
```

```python
# set up masks for carving
def v_mask(cost, path, size=1):
    """
        input: cost matrix H*W, path matrix H*W
        output: mask matrix H*W
        
        Calculate mask for vertical seam.
    """
    
    H, W = cost.shape
    mask = np.ones((H,W),dtype=bool) # initialize mask matrix
    if size == 1:
        idx = np.argmin(cost[-1,:])
        for i in range(H-1,0,-1):
            mask[i,idx] = False
            idx += path[i,idx]
        mask[0,idx] = False
    elif size > 1: # used only to demonstrate a BAD EXAMPLE
        idx = np.argsort(cost[-1,:])[:size]
        for j in idx:
            for i in range(H-1,0,-1):
                mask[i,j] = False
                j += path[i,j]
            mask[0,j] = False
            
    return mask

def h_mask(cost, path):
    """
        input: cost matrix H*W, path matrix H*W
        output: mask matrix H*W
        
        Calculate mask for horizontal seam.
    """
    
    H, W = cost.shape
    mask = np.ones((H,W),dtype=bool) # initialize mask matrix
    idx = np.argmin(cost[:,-1])
    for j in range(W-1,0,-1):
        mask[idx,j] = False
        idx += path[idx,j]
    mask[idx,0] = False
    
    return mask
```

## Step 3: Generate the result image

Now we just need to take the input image, define the dimension of the output image, and take out vertical and horizontal seams iteratively.

List of images:

    boat: https://pixabay.com/photos/lake-boat-fisher-man-landscape-5000642/
    island: https://www.peoplemattersglobal.com/article/others/no-corporation-is-an-island-23466

```python
im = cv2.imread("samples/boat.jpg")
plt.imshow(im[:,:,[2,1,0]])
```

```python
im.shape
```

```python
# show the first vertical seam and the first horizontal seam
seams = copy.deepcopy(im)

e = energy(im)

# vertical seam
cost, path = v_cost(e)
mask = v_mask(cost,path)
seams[:,:,2][~mask] = 255
seams[:,:,1][~mask] = 0
seams[:,:,0][~mask] = 0

# horizontal seam
cost, path = h_cost(e)
mask = h_mask(cost,path)
seams[:,:,2][~mask] = 255
seams[:,:,1][~mask] = 0
seams[:,:,0][~mask] = 0

plt.imshow(seams[:,:,[2,1,0]])
```

```python
# set up output dimension
output_height = 301
output_width = 574

input_height, input_width, input_channel = im.shape # input shape

# for video specifically - width and height divisible by 2
video_height = (input_height+1)//2*2
video_width = (input_width+1)//2*2
black_im = np.zeros((video_height,video_width,input_channel)) # align photos for video

v_cuts = input_width - output_width
h_cuts = input_height - output_height

ims = [im]

# carve out vertical seams first
for i in range(v_cuts):
    if (i+1)%10 == 0:
        print("v_cuts: {}/{}".format(i+1,v_cuts))
    e = energy(ims[-1])
    cost, path = v_cost(e)
    mask = np.repeat((v_mask(cost,path))[:,:,np.newaxis],input_channel,axis=2) # create mask for all channels
    ims.append(ims[-1][mask].reshape((input_height,input_width-1-i,input_channel)))

# carve out horizontal seams next
for i in range(h_cuts):
    if (i+1)%10 == 0:
        print("h_cuts: {}/{}".format(i+1,h_cuts))
    e = energy(ims[-1])
    cost, path = h_cost(e)
    mask = np.repeat((h_mask(cost,path))[:,:,np.newaxis],input_channel,axis=2) # create mask for all channels
    ims.append((ims[-1].transpose(1,0,2)[mask.transpose(1,0,2)]
                .reshape((output_width,input_height-1-i,input_channel))).transpose(1,0,2))
    
# show the result
plt.imshow(ims[-1][:,:,[2,1,0]])
```

```python
# create video
vids = []
for i in range(len(ims)):
    temp_im = copy.deepcopy(black_im)
    H,W,C = ims[i].shape
    temp_im[:H,:W,:C] += ims[i]
    vids.append(temp_im)
vids = np.array(vids,dtype=np.uint8)

utils.vidwrite_from_numpy('res_island_e2.mp4', vids[:,:,:,[2,1,0]], framerate=30)
```

## Extension 1: Seam Insertion

Seam insertion is like the "inverse" of seam carving. We identify the optimal seam and duplicate it once to the right. Note here that in order to not duplicate the same seam again and again, we identify all seams needed to enlarge the image first, and then duplicate each seam once.

```python
def idx_extension_orig_ext(ids):
    """
        input: dim n, starting idx's of the paths [original img] (at the bottom or at the right)
        output: dim n, according starting idx's for seam insertion (after the more optimal seams are inserted)
        
        from original image indexes to enlarged image indexes
        
    """
    res = np.zeros(len(ids),dtype=np.int32)
    for i in range(len(ids)):
        cnt = 0
        for j in range(i):
            if ids[j] < ids[i]:
                cnt += 1
        res[i] = ids[i] + cnt
    return res

def idx_extension_carve_orig(ids,size):
    """
        input: 
            ids: dim n, starting idx's of the paths [carved img] (at the bottom or at the right)
            size: size of the selected dimension of the [origimal img]
        output: dim n, according starting idx's for seam insertion (after the more optimal seams are inserted)
        
        from carved image indexes to original image indexes
        
    """
    res = []
    temp = list(range(size))
    for i in ids:
        res.append(temp[i])
        del temp[i]
    return res
```

```python
# Desired image size
output_height = 457
output_width = 690

input_height, input_width, input_channel = im.shape # input shape

# for video specifically - width and height divisible by 2
video_height = (output_height+1)//2*2
video_width = (output_width+1)//2*2
black_im = np.zeros((video_height,video_width,input_channel)) # align photos for video

v_inserts = output_width - input_width
h_inserts = output_height - input_height

ims = [im]

# insert vertical seams first

ims_forward = [ims[-1].astype(np.uint8)] # stores the ims in the forward pass (cut)
paths = [] # stores the paths from the forward pass
idxs = [] # stores the starting idx's of the paths

# step 1: carve out vertical seams 
for i in range(v_inserts):
    if (i+1)%10 == 0:
        print("v_insert forward pass: {}/{}".format(i+1,v_inserts))
    e = energy(ims_forward[-1])
    cost, path = v_cost(e)
    mask = np.repeat((v_mask(cost,path))[:,:,np.newaxis],input_channel,axis=2) # create mask for all channels
    ims_forward.append(ims_forward[-1][mask].reshape((input_height,input_width-1-i,input_channel)))
    paths.append(path)
    idxs.append(np.argmin(cost[-1,:]))

# step 2: map different index systems
v_bot_idx = idx_extension_carve_orig(idxs,input_width)
v_bot_idx_new = idx_extension_orig_ext(v_bot_idx)

# step 3: insert vetical seams into original image
for i in range(v_inserts):
    if (i+1)%10 == 0:
        print("v_inserts: {}/{}".format(i+1,v_inserts))
    im_new = copy.deepcopy(ims[-1]) # container for new image
    H,W,C = im_new.shape # shape of new image
    im_new = np.hstack((im_new,np.zeros((H,1,C),dtype=np.int32))) # expand size by 1
    idx = v_bot_idx_new[i]
    gap = v_bot_idx_new[i] - idxs[i] # gap between idx after insertion and idx before any insertion
    for j in range(H-1,-1,-1):
        im_new[j,idx+2:,:] = im_new[j,idx+1:-1,:]
        im_new[j,idx+1,:] = (im_new[j,idx,:] + (im_new[j,idx,:] if idx==W-1 else im_new[j,idx+2,:]))//2
        idx += paths[i][j,idx-gap]
    ims.append(im_new)

# insert horizontal seams next - INNOVATION

ims_forward = [ims[-1].astype(np.uint8)] # stores the ims in the forward pass (cut)
paths = [] # stores the paths from the forward pass
idxs = [] # stores the starting idx's of the paths

# step 1: carve out horizontal seams 
for i in range(h_inserts):
    if (i+1)%10 == 0:
        print("h_insert forward pass: {}/{}".format(i+1,h_inserts))
    e = energy(ims_forward[-1])
    cost, path = h_cost(e)
    mask = np.repeat((h_mask(cost,path))[:,:,np.newaxis],input_channel,axis=2) # create mask for all channels
    ims_forward.append((ims_forward[-1].transpose(1,0,2)[mask.transpose(1,0,2)]
                .reshape((output_width,input_height-1-i,input_channel))).transpose(1,0,2))
    paths.append(path)
    idxs.append(np.argmin(cost[:,-1]))

# step 2: map different index systems
h_right_idx = idx_extension_carve_orig(idxs,input_height)
h_right_idx_new = idx_extension_orig_ext(h_right_idx)

# step 3: insert vetical seams into original image
for i in range(h_inserts):
    if (i+1)%10 == 0:
        print("h_inserts: {}/{}".format(i+1,h_inserts))
    im_new = copy.deepcopy(ims[-1]) # container for new image
    H,W,C = im_new.shape # shape of new image
    im_new = np.vstack((im_new,np.zeros((1,W,C),dtype=np.int32))) # expand size by 1
    idx = h_right_idx_new[i]
    gap = h_right_idx_new[i] - idxs[i] # gap between idx after insertion and idx before any insertion
    for j in range(W-1,-1,-1):
        im_new[idx+2:,j,:] = im_new[idx+1:-1,j,:]
        im_new[idx+1,j,:] = (im_new[idx,j,:] + (im_new[idx,j,:] if idx==H-1 else im_new[idx+2,j,:]))//2
        idx += paths[i][idx-gap,j]
    ims.append(im_new)
    
# show the result
plt.imshow(ims[-1][:,:,[2,1,0]])
```

## Extension 2: Seam Carving with Forward Energy

Michael Rubinstein, Ariel Shamir, and Shai Avidan. 2008. Improved seam carving for video retargeting. <i>ACM Trans. Graph.</i> 27, 3 (August 2008), 1–9. https://doi.org/10.1145/1360612.1360615

Simple seam carving would likely result in some artifacts. Here we take the after-carving energy into consideration to guide the carving.

```python
# set up dynamic programming matrix for forward energy

def v_cost_forward(im):
    """
        input: im, H*W*3 (image with 3 color channels)
        output: 
            cost matrix, H*W. Cumulative forward energy cost from top to bottom.
            path matrix, H*W. Each cell records the RELATIVE column index of the prior pixel in the row above
                              (except for cells in the first row).
        
        Calculates the minimum cumulative forward energy cost of vertical paths and records the optimal paths.
    """
    H,W,C = im.shape
    cost = np.zeros((H,W)) # initialize the cost matrix
    path = np.zeros((H,W),dtype=np.int32) # initialize the path matrix
    
    # calculate shifted images
    im_u = np.roll(im,1,axis=0)
    im_l = np.roll(im,1,axis=1)
    im_r = np.roll(im,-1,axis=1)
    
    # initialize cX matrices
    cU = np.sum(np.abs(im_r-im_l),axis=2)
    cL = cU + np.sum(np.abs(im_u-im_l),axis=2)
    cR = cU + np.sum(np.abs(im_u-im_r),axis=2)
    
    cost[0,:] = cU[0,:]
    
    for i in range(1,H):
        for j in range(W):
            if j == 0: # edge case: left-most pixel
                cost_opts = [cost[i-1,j]+cU[i,j],
                             cost[i-1,j+1]+cR[i,j]]
                cost[i,j] = np.min(cost_opts)
                path[i,j] = np.argmin(cost_opts)
            elif j == W-1: # edge case: right-most pixel
                cost_opts = [cost[i-1,j-1]+cL[i,j],
                             cost[i-1,j]+cU[i,j]]
                cost[i,j] = np.min(cost_opts)
                path[i,j] = np.argmin(cost_opts)-1
            else:
                cost_opts = [cost[i-1,j-1]+cL[i,j],
                             cost[i-1,j]+cU[i,j],
                             cost[i-1,j+1]+cR[i,j]]
                cost[i,j] = np.min(cost_opts)
                path[i,j] = np.argmin(cost_opts)-1            
                
    return cost, path

def h_cost_forward(im):
    """
        TODO
    """
    cost_T, path_T = v_cost_forward(im.transpose(1,0,2))
    cost = cost_T.T
    path = path_T.T
    return cost, path
```

```python
# set up output dimension
output_height = 377
output_width = 590

input_height, input_width, input_channel = im.shape # input shape

# for video specifically - width and height divisible by 2
video_height = (input_height+1)//2*2
video_width = (input_width+1)//2*2
black_im = np.zeros((video_height,video_width,input_channel)) # align photos for video

v_cuts = input_width - output_width
h_cuts = input_height - output_height

ims = [im]

# carve out vertical seams first
for i in range(v_cuts):
    if (i+1)%10 == 0:
        print("v_cuts: {}/{}".format(i+1,v_cuts))
    cost, path = v_cost_forward(ims[-1])
    mask = np.repeat((v_mask(cost,path))[:,:,np.newaxis],input_channel,axis=2) # create mask for all channels
    ims.append(ims[-1][mask].reshape((input_height,input_width-1-i,input_channel)))

# carve out horizontal seams next
for i in range(h_cuts):
    if (i+1)%10 == 0:
        print("h_cuts: {}/{}".format(i+1,h_cuts))
    cost, path = h_cost_forward(ims[-1])
    mask = np.repeat((h_mask(cost,path))[:,:,np.newaxis],input_channel,axis=2) # create mask for all channels
    ims.append((ims[-1].transpose(1,0,2)[mask.transpose(1,0,2)]
                .reshape((output_width,input_height-1-i,input_channel))).transpose(1,0,2))
    
# show the result
plt.imshow(ims[-1][:,:,[2,1,0]])
```

```python
# create video
vids = []
for i in range(len(ims)):
    temp_im = copy.deepcopy(black_im)
    H,W,C = ims[i].shape
    temp_im[:H,:W,:C] += ims[i]
    vids.append(temp_im)
vids = np.array(vids,dtype=np.uint8)

utils.vidwrite_from_numpy('res_boat_fwd_e2.mp4', vids[:,:,:,[2,1,0]], framerate=30)
```
